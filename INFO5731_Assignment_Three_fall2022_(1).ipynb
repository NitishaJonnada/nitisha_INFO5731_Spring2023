```python
import nltk
from nltk import ngrams
import pandas as pd

reviews = pd.read_csv('Twitter_Reviews.csv')

# Check the actual column names
print(reviews.columns)

# Define the size of the n-grams
n = 3

# Create an empty dictionary to store the counts of each trigram
trigram_counts = {}

# Loop over each review and count the trigrams
for review in reviews['reviews']:
    tokens = nltk.word_tokenize(review.lower())
    trigrams = ngrams(tokens, n)
    for trigram in trigrams:
        if trigram in trigram_counts:
            trigram_counts[trigram] += 1
        else:
            trigram_counts[trigram] = 1

# Print the trigram counts
for trigram, count in trigram_counts.items():
    print(trigram, count)
```

```python
import nltk
from nltk import ngrams

reviews = pd.read_csv('Twitter_Reviews.csv')

# Define the size of the n-grams
n = 2

# Create an empty dictionary to store the counts of each bigram
bigram_counts = {}

# Loop over each review and count the bigrams
for review in reviews['reviews']:
    tokens = nltk.word_tokenize(review.lower())
    bigrams = ngrams(tokens, n)
    for bigram in bigrams:
        if bigram in bigram_counts:
            bigram_counts[bigram] += 1
        else:
            bigram_counts[bigram] = 1

# Calculate the probabilities for each bigram
for bigram, count in bigram_counts.items():
    w2, w1 = bigram
    w2_count = sum([v for k, v in bigram_counts.items() if k[0] == w2])
    probability = count / w2_count
    print(bigram, probability)
```

```python
import nltk
from nltk import ngrams
from textblob import TextBlob

reviews = pd.read_csv('Twitter_Reviews.csv')

# Create an empty dictionary to store the counts of each noun phrase
noun_phrase_counts = {}

# Loop over each review and count the noun phrases
for review in reviews['reviews']:
    blob = TextBlob(review.lower())
    noun_phrases = blob.noun_phrases
    for noun_phrase in noun_phrases:
        if noun_phrase in noun_phrase_counts:
            noun_phrase_counts[noun_phrase] += 1
        else:
            noun_phrase_counts[noun_phrase] = 1

# Calculate the relative probabilities for each noun phrase in each review
for i, review in reviews.iterrows():
    blob = TextBlob(review['reviews'].lower())
    noun_phrases = blob.noun_phrases
    if len(noun_phrases) == 0:
        continue
    max_count = max([noun_phrase_counts[noun_phrase] for noun_phrase in noun_phrases if noun_phrase in noun_phrase_counts])
    for noun_phrase in noun_phrases:
        if noun_phrase not in noun_phrase_counts:
            continue
        frequency = noun_phrase_counts[noun_phrase] / max_count
        print(review['reviews'], noun_phrase, frequency)
```

# **Question 2: Undersand TF-IDF and Document representation**

(20 points). Starting from the documents (all the reviews, or abstracts, or tweets) collected for assignment two, write a python program: 

(1) To build the **documents-terms weights (tf*idf) matrix bold text**.

(2) To rank the documents with respect to query (design a query by yourself, for example, "An Outstanding movie with a haunting performance and best character development") by using **cosine similarity**.

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Read the csv file containing the reviews
reviews = pd.read_csv('Twitter_Reviews.csv')

# Define a query
query = "An Outstanding movie with a haunting performance and best character development"

# Create a TfidfVectorizer object with stopwords and English language
tfidf_vectorizer = TfidfVectorizer(stop_words='english', lowercase=True)

# Fit and transform the reviews dataset to the vectorizer
tfidf_matrix = tfidf_vectorizer.fit_transform(reviews['reviews'])

# Calculate cosine similarity between the query and each document
query_tfidf = tfidf_vectorizer.transform([query])
cosine_similarities = cosine_similarity(query_tfidf, tfidf_matrix).flatten()

# Sort the documents by their cosine similarity to the query
document_scores = [(score, index) for index, score in enumerate(cosine_similarities)]
document_scores.sort(reverse=True)

# Print the top 10 documents and their scores
for score, index in document_scores[:10]:
    print(f"Document {index}: {reviews.iloc[index]['reviews']} (score: {score})")
```

# **Question 3: Create your own word embedding model**

(20 points). Use the data you collected for assignment two to build a word embedding model: 

(1) Train a 300-dimension word embedding (it can be word2vec, glove, ulmfit, bert, or others).

(2) Visualize the word embedding model you created.

Reference: https://machinelearningmastery.com/develop-word-embeddings-python-gensim/

Reference: https://jaketae.github.io/study/word2vec/

```python
# Write your code here

import pandas as pd
import gensim
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Read the csv file containing the reviews
reviews = pd.read_csv('Twitter_Reviews.csv')

# Tokenize the reviews
tokenized_reviews = [review.split() for review in reviews['reviews']]

# Train a Word2Vec model
model = Word2Vec(tokenized_reviews, vector_size=300, window=5, min_count=5, workers=4)

# Visualize the word embedding
words = list(model.wv.index_to_key)
X = model.wv[words]
pca = PCA(n_components=2)
result = pca.fit_transform(X)

plt.scatter(result[:, 0], result[:, 1])
for i, word in enumerate(words):
    plt.annotate(word, xy=(result[i, 0], result[i, 1]))
plt.show()

```
